```python
import os
import pinecone
from dotenv import load_dotenv
load_dotenv(dotenv_path='../.env')

# get api key from app.pinecone.io for .env file
lines = open('../.env').read().splitlines()
for line in lines:
    if line.startswith('PINECONE_API_KEY'):
        api_key = line.split('=')[1]
        break
api_key = api_key or os.environ.get('PINECONE_API_KEY')
# find your environment next to the api key in pinecone console
env = os.environ.get('PINECONE_ENVIRONMENT') or 'gcp-starter'

pinecone.init(
    api_key=api_key,
    environment=env
)
```

```python
# list all indexes
indexes = pinecone.list_indexes()
indexes
```

```python
# delete an index
for index in indexes:
    pinecone.delete_index(index)
```

```python
from pinecone_datasets import list_datasets, load_dataset

dataset = load_dataset("ANN_Fashion-MNIST_d784_euclidean")
```

```python
pinecone.create_index(name="fashion-mnist", dimension=784, metric="euclidean")
```

```python

dataset.to_pinecone_index('fashion-mnist',api_key=api_key,environment=env,should_create_index=False,batch_size=300)
```
